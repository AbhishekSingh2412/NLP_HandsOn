```python
#BASIC TEXT PROCESSING PIPELINE
import nltk
text1 = ''' The only true wisdom is knowing that you know nothing. 
This classic quote by Socrates means that only person who thinks that 
he knows nothing will try to achieve knowledge 
from anyone and everyone he comes across in 
life then that person might 
be a king or a beggar. '''

for text in text1:
    sentences=nltk.sent_tokenize(text1)
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
        tagged = nltk.pos_tag(words)
        print(tagged)
```

```python
#TASK 2: Tweet Tokenizer

from nltk.tokenize import TweetTokenizer
text = 'The Part was sooo fun :D #superfun'
twtkn= TweetTokenizer()
twtkn.tokenize(text)
```

```python
#COMPARISON between tweet tokenize and word tokenize
words=nltk.word_tokenize('The Part was sooo fun :D #superfun')
print(words)
```
